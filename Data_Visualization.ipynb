```python
import seaborn as sns # Data Visualization
import pandas as pd # Data Manipulation
```

```python
# Import dataset

df = sns.load_dataset("tips")
```

```python
df.head(10) # Menampilkan top 5 baris data pertama
```

```python
# Count Plot -- count the number of cateogical data classes
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(x = "day",
              data = df)
plt.title("Count of Tips by Day in Week")
plt.show()
```

```python
# Box Plot --> visualize distribution of numerical within categories groups
plt.figure(figsize=(8,5))
sns.boxplot(x = "time",
            y = "total_bill",
            data = df)

plt.title("Total Bill Distribution by Time")
plt.show()
```

```python
# Violin Plot

plt.figure(figsize=(8,5))
sns.violinplot(x = "time",
            y = "total_bill",
            data = df)

plt.title("Total Bill Distribution by Time")
plt.xlabel("Meal Time")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
# Distplot
sns.distplot(a = df["total_bill"])
plt.title("Total Bill Distribution")
plt.show()
```

```python
#Facetgrid --> feature allowing to create a grid of subplots by
#combinations of different subset

f = sns.FacetGrid(df, col="day", height = 6)
f.map(sns.distplot, "total_bill")
f.set_axis_labels("Total Bill", "Frequency")
plt.show()
```

```python
# Correlation/scatter plot - Numerical to numerical
# Regression Plot
sns.regplot(x = "total_bill",
            y = "tip",
            data = df,
            line_kws = {"color":"red"})
plt.title("Correlation of Total Bill with Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

# MultiVariate Plots

```python
sns.violinplot(x="time",
               y = "total_bill",
               data = df,
               hue = "sex",
               split = True)

plt.title("Violin Plot with HUe for Total Bill by Time and Gender")
plt.xlabel("Time")
plt.ylabel("Total Bill")
plt.legend(title = "Gender")
plt.savefig("Violin Plot by Time and Gender with Total Bill.png")
plt.show()
```

```python
sns.scatterplot(x = "total_bill",
                y = "tip",
                data = df,
                hue = "day",
                style = "sex") # hue = warna, style = lambang

plt.title("Scatter Plot with Days and Tips")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.legend()
plt.show()
```
